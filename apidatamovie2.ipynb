# TO REQUEST A SINGLE PAGE

```python
response=requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=4a3b875418abdb6df6b55006e13d9fed&page=1') ####sent https request to fetch the data through api and stored in response object.
```

```python
response.json() ###has key value pair {}
```

```python
response.json()['total_pages']
```

```python
response.json()['results'] ##have all the movie data in different dictionary stored  in a list
```

```python
resultsdata=response.json()['results']
```

```python
resultsdata ###it is a list having multiple dictionaries as element.
```

## ACCESSING PAGE 1 DATA AND CREATING A DATAFRAME

```python
id=[]
title=[]
release_date=[]
popularity=[]
vote_average=[]
vote_count=[]
    
```

```python
for i in resultsdata:
        
        id.append(i['id'])
        title.append(i['title'])
        
        release_date.append(i['release_date'])
        
        
        popularity.append(i['popularity'])
    
        vote_average.append(i['vote_average'])
        
        vote_count.append(i['vote_count'])
    
        
```

```python
len(popularity)
```

```python
len(title)
```

```python
dataset=pd.DataFrame({"id":id,"title":title,"popularity":popularity,"release_date":release_date,"vote_average":vote_average,"vote_count":vote_count})
```

# ACCESSING ALL 501 PAGE AND CREATING DATAFRAME

```python
id1=[]
title1=[]
release_date1=[]
popularity1=[]
vote_average1=[]
vote_count1=[]
    
for i in range(1,501):
    response1=requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=4a3b875418abdb6df6b55006e13d9fed&page={}'.format(i))
    resultdata=response1.json()['results']
    for i in resultdata:
        
        id1.append(i['id'])
        title1.append(i['title'])
        
        release_date1.append(i['release_date'])
        
        
        popularity1.append(i['popularity'])
    
        vote_average1.append(i['vote_average'])
        
        vote_count1.append(i['vote_count'])
        
    
dataset1=pd.DataFrame({"id":id1,"title":title1,"popularity":popularity1,"release_date":release_date1,"vote_average":vote_average1,"vote_count":vote_count1}) 
                           
```
